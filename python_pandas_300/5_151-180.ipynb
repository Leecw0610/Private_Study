```python
data1 = {
    'ID': [100, 101, 102, 103],
    'Name': ['Brayden', 'Clara', 'Ella', 'Oliver']
}

data2 = [
    [100, "삼성전자"],
    [100, "LG전자"],
    [101, "카카오"],
    [101, "NAVER"],
    [101, "현대차"]
]

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data=data2, columns=['ID', '종목명'])
```

```python
df1
```

```python
df2 
```

```python
pd.concat([df1,df2],join='inner',axis=0)
```

```python
pd.merge(df1, df2, how='inner', on='ID')
```

```python
pd.merge(df1, df2, how='outer', on='ID')
```
