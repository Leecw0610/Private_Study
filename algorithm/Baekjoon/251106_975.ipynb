```python
from sys import stdin
from collections import deque
input = stdin.readline

N, K = map(int, input().split())

def bfs(start):
    max = 100000 # 최대 입력 수
    q = deque()
    q.append((start, 0)) #시작점, 시간 기본 0
    visited = set() #방문여부, 한번 간곳은 X

    while q:
        v, time = q.popleft()

        # v가 K랑 같아지면 시간 출력
        if v == K:
            return time

        #방문 표시
        if v not in visited:
            visited.add(v)
            # 범위별 1초씩 추가하면서 모든 수 저장
            if v-1 >= 0:
                q.append((v-1, time+1))
            if v+1 <= max:
                q.append((v+1, time+1))
            if v*2 <= max:
                q.append((v*2, time+1))

print(bfs(N))
```
