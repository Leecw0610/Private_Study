```python
r, c = map(int, input().split())
l = []
for i in range(r):
    l.append(list(input()))
    
word = []
for i in range(r):
    w = ''
    for j in range(c):
        if l[i][j] == '#':
            if w != '' and len(w) > 1:
                word.append(w)
            w = ''
        else :
            w += l[i][j]
    if w != '' and len(w) > 1:
        word.append(w)
    
for i in range(c):
    w = ''
    for j in range(r):
        if l[j][i] == '#':
            if w != '' and len(w) > 1:
                word.append(w)
            w = ''
        else :
            w += l[j][i]
    if w != '' and len(w) > 1:
        word.append(w)

print(sorted(word)[0])
```
