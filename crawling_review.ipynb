```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://kr.noxinfluencer.com/youtube-channel-rank/top-100-kr-all-youtuber-sorted-by-subs-weekly'
```

```python
req = requests.get(url)
html = req.text
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
soup.select('#table-body > div:nth-child(1) > span.rank-cell.pull-left.rank-desc > a > span')
```

```python
soup.select('#table-body > div:nth-child(2) > span.rank-cell.pull-left.rank-desc > a > span')
```

```python
name = soup.select('#table-body > div > span.rank-cell.pull-left.rank-desc > a > span')
```

```python
name[0]
```

```python
name[0].text
```

```python
name_t =[]
for n in name :
    name_t.append(n.text)  
```

```python
name_t
```

```python
cate = soup.select('#table-body > div > span.rank-cell.pull-left.rank-category > a')
```

```python
cate_t =[]
for t in cate :
    cate_t.append(t.text)
```

```python
cate_t
```

```python
subs=soup.select('#table-body > div > span.rank-cell.pull-left.rank-subs > span.number')
```

```python
subs_t =[]
for t in subs :
    subs_t.append(t.text)
```

```python
subs_t
```

```python
watch =soup.select('#table-body > div > span.rank-cell.pull-left.rank-avg-view > span.number')
```

```python
watch_t =[]
for t in watch :
    watch_t.append(t.text)
```

```python
watch_t
```

```python
data  = list(zip(name_t, cate_t, subs_t, watch_t))
```

```python
df = pd.DataFrame(data, columns=['이름','카테고리','구독자','평균조회수'])
```

```python
df
```

```python
df['구독자']
```

```python
for i in range (100):
    df['구독자'][i]= int(float(df['구독자'][i].strip('만')) *10000)
    i+= 1
```

```python
for i in range (100):
    df['평균조회수'][i]= int(float(df['평균조회수'][i].strip('만')) *10000)
    i+= 1
```
