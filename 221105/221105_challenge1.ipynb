```python
PageNo = int(input('페이지를 입력하세요.'))

data=[]

for n in tqdm(range(PageNo)) :
    
    #print(n)
    url = 'https://socialerus.com/Index?PageNo='+str(n+1)+'&ChCategory=&ChTheme=&ChSubscMin=&ChSubscMax=&ChViewMin=&ChViewMax=&ChVideoMin=&ChVideoMax=&Sort=subsc'
    # 페이지에 맞는 url변경!
    req = requests.get(url)
    html = req.text
    soup = BeautifulSoup(html, 'html.parser')
    inform=soup.select('#content > div.ranking_wrap > div:nth-child(2) > div > div.ranking_cont > ul.list_ranking > li')
    
    for i in range(len(inform)) :
        #print(i)
        temp=[]
        temp.append((20*n)+(int(i)+1)) #순위
        temp.append(inform[i].find('span', {'class':'title'}).text) #채널이름
        temp.append(inform[i].find('span', {'class':'category'}).text) #장르
        temp.append(int(inform[i].find('span', {'class':'num subscriber'}).text.replace(',',''))) #구독자수
        temp.append(int(inform[i].find('span', {'class':'num views'}).text.replace(',',''))) #조회수
        temp.append(int(inform[i].find('span', {'class':'num video'}).text.replace(',',''))) #영상수
        data.append(temp)
```

```python
data
```

```python
df=pd.DataFrame(data, columns=['순위', '채널이름', '장르', '구독자수', '조회수', '영상수'])
```
