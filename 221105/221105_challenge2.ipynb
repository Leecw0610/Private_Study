```python
data=[]
n=int(input('순위 범위를 입력하세요')) #n은 순위

for i in tqdm(range(n)) :
    element=soup.select('#search_resultsRows > a')
    title=element[i].find('span', {'class':'title'}).text
    try :
        price=int((element[i].find('div' , {'class' :'col search_price_discount_combined responsive_secondrow'}).text.replace(',','').replace('₩','').strip())[-6:])
    except :
        price=0
    try : 
        discount=element[i].find('div',{'class':'col search_discount responsive_secondrow'}).text.replace('\n','').replace('%','')
    except :
        discount=0
    #print(i)
    temp=[]
    temp.append(i+1)
    temp.append(title)
    temp.append(price)
    temp.append(discount)
    data.append(temp)
```

```python
data
```

```python
df=pd.DataFrame(data, columns=['순위','게임명','가격','할인율'])
```
