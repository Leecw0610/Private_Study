```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/chart/boxoffice/?ref_=nv_ch_cht'
```

```python
req = requests.get(url)
html = req.text
soup= BeautifulSoup(html, 'html.parser')
```

```python
soup.select('#boxoffice > table > tbody > tr > td>td.titleColumn > a')
```

```python
title=soup.select('#boxoffice > table > tbody > tr > td>td.titleColumn > a')
```

```python
title[0]['href']
```

```python
data = []
for t in title :
    #print(t.text)
    temp =[]
    temp.append(t.text)
    temp.append(t['href'].replace('/title/',''))
    data.append(temp)
```

```python
data
```

```python
df = pd.DataFrame(data, columns= ['영화명','코드'])
```

```python
def soup_get(url) : 
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'html.parser')
    return soup
```

```python
for code in df['코드'][:1] :
    url2 = 'https://www.imdb.com/title/' + code + '/reviews?spoiler=hide&sort=curated&dir=desc&ratingFilter=0'
    soup = soup_get(url2)
    #print(soup)
    review= soup.select('#main > section > div.lister > div.lister-list > div')
```

```python
import tqdm as tqdm
data1=[]
for code in df['코드']:
    #print(code)
    url2 = 'https://www.imdb.com/title/' + code + '/reviews?spoiler=hide&sort=curated&dir=desc&ratingFilter=0'
    soup = soup_get(url2)
    review= soup.select('#main > section > div.lister > div.lister-list > div')
    
    for i in review :
        temp=[]
        temp.append(code)
        temp.append(i.find('a',{'class':'title'}).text.strip())
        try: 
            temp.append(int(i.find('span').text.strip().replace('/10','')))
        except:
            temp.append('Empty')
            #try except를 이용한 empty값의 예외처리!
        temp.append(i.find('div',{'class':'text show-more__control'}).text.strip())
        data1.append(temp)
```
